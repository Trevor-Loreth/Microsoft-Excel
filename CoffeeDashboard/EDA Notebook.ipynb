# Exploratory Data Analysis
Importing the raw data to understand the columns and data available to me and to determine the vision of my dashboard.

```python
import pandas as pd

# Importing file
file_path = "Coffee Dashboard.xlsx"
df = pd.read_excel(file_path, sheet_name="Raw Data")

df
```

```python
print("\nShape:", df.shape) # (row, col)
print("\nColumns:", df.columns)
```

```python
print("\nData Types")
print(df.dtypes)

# Looking for any null values in the columns I plan to use
print("\nMissing Values")
print(df.isnull().sum())
```

```python
# Checking for duplicate rows
print(df.duplicated().sum())
```

```python
# Replacing the original date time with the datetime_col
datetime_col = "datetime"

# Converting column to datetime
df[datetime_col] = pd.to_datetime(df[datetime_col])

# Creating a time only column in HH:MM:SS AM/PM format
df["Time"] = df[datetime_col].dt.strftime("%I:%M:%S %p")

# Finding the earliest and latest time
earliest_time = df[datetime_col].dt.time.min()
latest_time = df[datetime_col].dt.time.max()

print("Earliest time:", earliest_time)
print("Latest time:", latest_time)
```

```python
# Comparing Time and datetime
print(df[[datetime_col, "Time"]].head())
```

```python
# Stat summary
df.describe()
```

```python
# Checking top coffee purchases
print(df["coffee_name"].value_counts())
```
